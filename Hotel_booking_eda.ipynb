# Data Analysis and Visualizations

The below bar graph shows the percentage of reservations that are cancelled
and those that are not. 

```python
cancelled_percentage = df['is_canceled'].value_counts(normalize = True).mul(100).round(2).astype(str) + "%"
print(cancelled_percentage)

plt.figure(figsize = (5,4))
plt.title('Reservation status count',size=20)
ax=plt.bar(['Not_canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.5)
plt.xlabel("Cancellation status")
plt.ylabel("No. of reservations")
plt.bar_label(ax)
plt.show()
```

Reservation status in different hotels

```python
plt.figure(figsize = (5,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'summer')
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('No. of reservations')
plt.legend(['not canceled', 'canceled'])
for bars in ax1.containers:
    plt.bar_label(bars)
plt.show()
```

Absolute and percentage wise cancellations of both hotels

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
pd.concat([resort_hotel['is_canceled'].value_counts(normalize = True).mul(100).round(2).astype(str) + "%", resort_hotel['is_canceled'].value_counts()], axis=1)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
pd.concat([city_hotel['is_canceled'].value_counts(normalize = True).mul(100).round(2).astype(str) + "%", city_hotel['is_canceled'].value_counts()], axis=1)
```

Is there any effect of price for reservation cancelation in different hotels

```python
resort_hotel_avg_adr_daywise=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
resort_hotel_avg_adr_daywise
```

```python
city_hotel_avg_adr_daywise = city_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_avg_adr_daywise
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel_avg_adr_daywise.index, resort_hotel_avg_adr_daywise['adr'], label = 'Resort Hotel')
plt.plot(city_hotel_avg_adr_daywise.index, city_hotel_avg_adr_daywise['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.ylabel("ADR")
plt.xlabel("Reservation Date")
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax2=sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
for bars in ax2.containers:
    plt.bar_label(bars)
plt.legend(['not canceled', 'canceled'])
plt.show()
```

To understand the reservation status (Canceled, Not_Canceled) and average price to know the price trend over time in both categories so that we can understand people behavior well

```python
cancelled_data= df[df['is_canceled'] == 1]
cancelled_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_adr.reset_index(inplace = True)
cancelled_adr.sort_values('reservation_status_date', inplace=True)
cancelled_adr
```

```python
not_cancelled_data= df[df['is_canceled'] == 0]
not_cancelled_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_adr.reset_index(inplace = True)
not_cancelled_adr.sort_values('reservation_status_date', inplace=True)
not_cancelled_adr
```

```python
cancelled_data= df[df['is_canceled'] == 1]
cancelled_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_adr.reset_index(inplace = True)
cancelled_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_adr.reset_index(inplace = True)
not_cancelled_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate for cancelled and not cancelled reservations', size=30)
sns.lineplot(x='reservation_status_date',y='adr', data=not_cancelled_adr)
sns.lineplot(x='reservation_status_date',y='adr', data=cancelled_adr)
plt.legend(['not cancelled', 'cancelled'])
plt.show()
```

```python
cancelled_adr_monthwise = cancelled_data.groupby('month')[['adr']].mean()
cancelled_adr_monthwise.reset_index(inplace = True)
cancelled_adr_monthwise.sort_values('month', inplace=True)

not_cancelled_adr_monthwise = not_cancelled_data.groupby('month')[['adr']].mean()
not_cancelled_adr_monthwise.reset_index(inplace = True)
not_cancelled_adr_monthwise.sort_values('month', inplace=True)

plt.figure( figsize = (20,8))
plt.title('Average Daily Rate in reservation status over month for canceled and not canceled ', fontsize = 20)
sns.lineplot(x='month',y='adr', data=not_cancelled_adr_monthwise)
sns.lineplot(x='month',y='adr', data=cancelled_adr_monthwise)
plt.legend(['not cancelled', 'cancelled'])
plt.show()
```
